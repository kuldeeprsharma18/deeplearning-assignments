In this assignment students have to transform iris data into 3 dimensions and plot a 3d chart with transformed dimensions and colour each data point with specific class.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
from sklearn import datasets
```

```python
%matplotlib inline
# importing Iris dataset from sklearn.datasets 
iris = datasets.load_iris()
X = iris.data[:, :2]  # taking only the first two features.
y = iris.target

# for getter a better understanding of interaction of the dimensions
# plotting the 3d chart with transformed dimensions using PCA
fig = plt.figure(1, figsize=(10, 8))
a = Axes3D(fig, elev=-150, azim=120)
X_reduced = PCA(n_components=3).fit_transform(iris.data)
a.scatter(X_reduced[:, 0], X_reduced[:, 1], X_reduced[:, 2],
           cmap=plt.cm.Set3, c=y, edgecolor='k', s=40)
a.set_title("3d chart with transformed dimensions")
a.set_xlabel("1st eigenvector")
a.set_ylabel("2nd eigenvector")
a.set_zlabel("3rd eigenvector")

plt.show()
```
